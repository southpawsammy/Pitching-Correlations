```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
bullpen = pd.read_csv('pitchingAVGBreaks.csv')
x = bullpen[["z_swing_miss_percent"]]
y = bullpen[["fastball_avg_break"]]

regr = LinearRegression()
regr.fit(x,y)
y_pred = regr.predict(x)
```

```python
plt.scatter(x, y, color = 'black')
plt.plot(x, regr.predict(x), color= 'red')
plt.title('Swing and Miss rates vs Fastball break AVG')
plt.xlabel('Swing and Miss (%)')
plt.ylabel('Fastball break AVG (inches)')
plt.show() 
```
