```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
bullpen = pd.read_csv('pitchingAVGBreaks.csv')
x = bullpen[["z_swing_miss_percent"]]
y = bullpen[["breaking_avg_break"]]
regr = LinearRegression()
regr.fit(x,y)
y_pred = regr.predict(x)


plt.scatter(x, y, color = 'black')
plt.plot(x, regr.predict(x), color= 'red')
plt.title('Swing and Miss rates vs Breaking ball break AVG')
plt.xlabel('Swing and Miss (%)')
plt.ylabel('Breaking ball break AVG (inches)')
plt.show() 
```
